# Bộ lọc thông thấp Gauss

```python
import numpy as np
import matplotlib.pyplot as plt

# ===  Tao bo loc Gauss trong mien khong gian ===
P, Q = 256, 256
sigma = 10
x = np.arange(P) - P/2
y = np.arange(Q) - Q/2
X, Y = np.meshgrid(x, y, indexing='ij')

h = np.exp(-(X**2 + Y**2) / (2 * sigma**2))
h = h / np.sum(h)

# === DICH TAM VE GOC truoc khi FFT ===
h_shifted = np.fft.ifftshift(h)

# === Bien doi Fourier ===
H = np.fft.fft2(h_shifted)
H = np.fft.fftshift(np.abs(H))
H = H / np.max(H)

# ===  Hien thi ket qua ===
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(h, cmap='gray')
plt.title('Bo loc Gauss trong mien khong gian')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(H, cmap='gray')
plt.title('Bo loc sau khi chuyen sang mien tan so')
plt.axis('off')

plt.tight_layout()
plt.show()
```
